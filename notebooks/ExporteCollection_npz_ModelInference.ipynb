```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

PIXEL_SIZE = 30
CHIP_SIZE = 256

@retry(tries=2, delay=5, backoff=3)
def downloadEachImage(args):
    """Download each file of list."""
    expression = args[0]
    region_dict = args[1]
    name = args[2]

    region = ee.Geometry(region_dict)
    proj = ee.Projection("EPSG:4326").atScale(PIXEL_SIZE).getInfo()
    scale_y = -proj["transform"][0]
    scale_x = proj["transform"][4]

    listCoords = ee.Array.cat(region.coordinates(), 1).getInfo()
    centroid = region.centroid(0.001).coordinates().getInfo()
    lon, lat = centroid[0], centroid[1]
    xMin = listCoords[0][0]
    yMax = listCoords[2][1]
    coords = [xMin, yMax]

    grid = {
        "dimensions": {"width": CHIP_SIZE, "height": CHIP_SIZE},
        "affineTransform": {
            "scaleX": scale_x,
            "scaleY": scale_y,
            "translateX": coords[0],
            "translateY": coords[1],
        },
        "crsCode": "EPSG:4326",
    }

    response = session.post(
        url=url,
        data=json.dumps(
            {
                "expression": ee.serializer.encode(expression),
                "fileFormat": "NPY",
                "bandIds": expression.bandNames().getInfo(),
                "grid": grid,
            }
        ),
    )

    try:
        error = json.loads(response.content)
        logger = logging.getLogger()
        logger.error(str(json.loads(response.content)))
        raise Exception("run again")
    except UnicodeDecodeError:
        array = np.load(io.BytesIO(response.content)).tolist()
        array = np.array(array)

        np.savez_compressed(
            f"/content/drive/MyDrive/ClayModel_detail/palm_ls/data/binaryclass/raw_test/{name}.npz",
            #f"/content/{name}.npz",
            array=array.transpose([2, 0, 1]),
            timestamp=expression.get('timestamp').getInfo(),
            lon=expression.get('longitude').getInfo(),
            lat=expression.get('latitude').getInfo(),
            gsd=expression.get('gsd').getInfo(),
        )

def parallel_download(tasks):
    """
    Execute downloadEachImage in parallel.

    Args:
        tasks: List of arguments for downloadEachImage.
        max_workers: Number of threads to use for parallel execution.
    """
    start_time = time.time()

    with ThreadPoolExecutor() as executor:
        future_to_task = {executor.submit(downloadEachImage, task): task for task in tasks}

        for future in as_completed(future_to_task):
            task = future_to_task[future]
            try:
                future.result()
                print(f"Task completed: {task[2]}")
            except Exception as exc:
                print(f"Task {task[2]} generated an exception: {exc}")

    end_time = time.time()
    print(f"Total execution time: {(end_time - start_time) / 60:.2f} minutes")
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time
import rasterio
import numpy as np
import matplotlib.pyplot as plt

PIXEL_SIZE = 30
CHIP_SIZE = 512
TARGET = "dende"
REGION = 'BA'
SENSOR = "landsat"
YEAR = "2012"

def plot_rgb_tiff(tiff_path):

    with rasterio.open(tiff_path) as src:
        print(src)
        # Lê as bandas 2, 1, 0 (considerando indexação 0)
        red = src.read(1)  # Banda 2
        green = src.read(2)  # Banda 1
        blue = src.read(3)  # Banda 0

        # Normaliza os valores para [0, 1] para visualização
        def normalize(band):
            band_min, band_max = band.min(), band.max()
            return (band - band_min) / (band_max - band_min + 1e-6)

        red_norm = red*2.75e-05
        green_norm = green*2.75e-05
        blue_norm = blue*2.75e-05

        # Empilha as bandas para formar a imagem RGB
        rgb = np.dstack((red_norm, green_norm, blue_norm))

        # Plotagem
        plt.figure(figsize=(10, 10))
        plt.imshow(rgb)
        plt.axis('off')
        plt.title('Imagem RGB (Bandas 2, 1, 0)')
        plt.show()


# Função de download para cada imagem
@retry(tries=2, delay=5, backoff=2)
def downloadEachImage(args):
    """Download each file of list."""
    expression = args[0]
    region_dict = args[1]
    name = args[2]

    region = ee.Geometry(region_dict)
    proj = ee.Projection("EPSG:4326").atScale(PIXEL_SIZE).getInfo()
    scale_y = -proj["transform"][0]
    scale_x = proj["transform"][4]

    listCoords = ee.Array.cat(region.coordinates(), 1).getInfo()
    centroid = region.centroid(0.001).coordinates().getInfo()
    lon, lat = centroid[0], centroid[1]
    xMin = listCoords[0][0]
    yMax = listCoords[2][1]
    coords = [xMin, yMax]

    grid = {
        "dimensions": {"width": CHIP_SIZE, "height": CHIP_SIZE},
        "affineTransform": {
            "scaleX": scale_x,
            "scaleY": scale_y,
            "translateX": coords[0],
            "translateY": coords[1],
        },
        "crsCode": "EPSG:4326",
    }

    response = session.post(
        url=url,
        data=json.dumps(
            {
                "expression": ee.serializer.encode(expression),
                "fileFormat": "GEO_TIFF",
                "bandIds": expression.bandNames().getInfo(),
                "grid": grid,
            }
        ),
    )

    try:
        error = json.loads(response.content)
        logger = logging.getLogger()
        logger.error(str(json.loads(response.content)))
        raise Exception("run again")
    except UnicodeDecodeError:
        # Caminho de saída para salvar o GeoTIFF
        geotiff_path = f"/content/drive/MyDrive/ClayModel_detail/palm_ls/data/predict/{TARGET}/{YEAR}/{REGION}/{name}.tif"

        # Salva o conteúdo da resposta diretamente como GeoTIFF
        with open(geotiff_path, "wb") as f:
            f.write(response.content)

        print(f"Arquivo GeoTIFF salvo em: {geotiff_path}")


def parallel_download(tasks):
    """
    Realiza o download das imagens em paralelo.

    Args:
        tasks: Lista de argumentos para a função downloadEachImage.
        max_workers: Número de threads para execução paralela.
    """
    with ThreadPoolExecutor() as executor:
        futures = {executor.submit(downloadEachImage, task): task for task in tasks}

        for future in as_completed(futures):
            task = futures[future]
            try:
                future.result()
                print(f"Download concluído para: {task[2]}")
            except Exception as exc:
                print(f"Erro no download para {task[2]}: {exc}")


def process_and_download_in_chunks(tiles_list, chunk_size=10):
    """
    Processa uma lista de imagens em blocos e baixa cada bloco imediatamente.

    Args:
        tiles_list: Lista de imagens do Earth Engine.
        chunk_size: Número de imagens por bloco a processar.
    """
    print(f"Número total de imagens: {len(tiles_list)}")

    for start in range(0, len(tiles_list), chunk_size):
        end = min(start + chunk_size, len(tiles_list))
        print(f"Processando imagens de {start} a {end - 1}")

        start_time = time.time()

        # Criar argumentos para o bloco atual
        tasks = []
        for i in range(start, end):
            image = tiles_list[i]  # Acessar diretamente a imagem na lista
            try:
                region_dict = image.get('system:footprint').getInfo()
            except Exception as e:
                print(f"Erro ao obter a geometria para a imagem {i}: {e}")
                continue  # Ignorar a imagem atual e continuar

            name = f'{SENSOR}{YEAR}_{i}'
            tasks.append((image, region_dict, name))

        # Fazer o download das tarefas do bloco atual
        if tasks:  # Apenas realizar download se houver tarefas válidas
            parallel_download(tasks)

        end_time = time.time()
        duration = (end_time - start_time) / 60
        print(f"Duração do processamento do bloco {start}–{end - 1}: {duration:.2f} min")

    print("Processamento concluído.")
```

```python
import os
from pathlib import Path
import rasterio
from rasterio.windows import Window
import numpy as np
import torch
import torch.nn.functional as F
from tqdm import tqdm
from torchvision.transforms import v2
import yaml

def get_model(chesapeake_checkpoint_path, clay_checkpoint_path, metadata_path):
    model = ChesapeakeSegmentor.load_from_checkpoint(
        checkpoint_path=chesapeake_checkpoint_path,
        metadata_path=metadata_path,
        ckpt_path=clay_checkpoint_path,
    )
    model.eval()
    return model

def create_transforms(mean, std):
    return v2.Compose([
        v2.Normalize(mean=mean, std=std),
    ])

def create_feathering_weight(height, width, overlap, feather_top=True, feather_bottom=True, feather_left=True, feather_right=True):
    """
    Cria uma matriz de pesos para blending suave nas bordas, aplicando o feathering somente
    para os lados internos. Se o feathering estiver desabilitado para uma borda (lado externo),
    os pesos nessa direção serão 1.

    Args:
        height, width (int): dimensões do chip.
        overlap (int): largura da região de sobreposição.
        feather_top, feather_bottom, feather_left, feather_right (bool): se True, aplica feathering para o lado correspondente.

    Retorna:
        weights: matriz de pesos com valores entre 0 e 1.
    """
    # Cria arrays para as coordenadas
    y = np.arange(height).reshape(-1, 1)
    x = np.arange(width).reshape(1, -1)

    # Se não for para aplicar feathering em um lado, forçamos a distância a ser igual ou maior que overlap (resultando em peso 1)
    top = y if feather_top else np.full((height, 1), overlap, dtype=np.int32)
    bottom = (height - y - 1) if feather_bottom else np.full((height, 1), overlap, dtype=np.int32)
    left = x if feather_left else np.full((1, width), overlap, dtype=np.int32)
    right = (width - x - 1) if feather_right else np.full((1, width), overlap, dtype=np.int32)

    # Calcula a distância mínima em relação a todas as bordas
    min_dist = np.minimum(np.minimum(top, bottom), np.minimum(left, right))

    # Converte essa distância em pesos lineares entre 0 e 1 na região de overlap
    weights = np.clip(min_dist / overlap, 0.0, 1.0).astype(np.float32)
    return weights

def compute_window_coords(image_dim, chip_size, stride):
    """
    Calcula as coordenadas de início para os patches ao longo de uma dimensão (largura ou altura)
    de modo que a última janela seja exatamente alinhada com a borda.

    Args:
        image_dim (int): dimensão da imagem (largura ou altura).
        chip_size (int): tamanho do patch.
        stride (int): passo de janelamento.

    Retorna:
        List[int]: lista de coordenadas de início.
    """
    coords = []
    # Caso a imagem seja menor ou igual ao patch, retorne apenas 0
    if image_dim <= chip_size:
        return [0]

    pos = 0
    while True:
        # Se o patch atual ultrapassa (ou chega exatamente em) a borda, force a última posição
        if pos + chip_size >= image_dim:
            coords.append(image_dim - chip_size)
            break
        else:
            coords.append(pos)
            pos += stride
    return coords



def split_geotiff(image_path, chip_size, overlap):
    """Divide a imagem em janelas (chips) com padding e sobreposição controlada, garantindo
    que, ao atingir a borda, não sejam gerados múltiplos patches sobre a mesma região."""
    with rasterio.open(image_path) as src:
        width, height = src.width, src.height
        meta = src.meta.copy()
        patches = []

        stride = chip_size - overlap

        # Calcula as coordenadas de início dos patches para cada dimensão
        x_coords = compute_window_coords(width, chip_size, stride)
        y_coords = compute_window_coords(height, chip_size, stride)

        for y in y_coords:
            for x in x_coords:
                # Define a janela
                win = Window(
                    col_off=x,
                    row_off=y,
                    width=chip_size,
                    height=chip_size
                )

                # Lê o patch (regiões fora da imagem serão preenchidas com 0)
                chip = src.read(window=win, boundless=True, fill_value=0).astype(np.float32)

                # Calcula a área válida do patch (caso o patch extrapole a imagem original)
                valid_width = chip_size if (x + chip_size) <= width else width - x
                valid_height = chip_size if (y + chip_size) <= height else height - y

                # Define flags para feathering:
                # Se o patch estiver na borda externa, desabilita o feathering naquela direção.
                feather_top = False if y == 0 else True
                feather_left = False if x == 0 else True
                feather_bottom = False if (y + chip_size) >= height else True
                feather_right = False if (x + chip_size) >= width else True

                # Gera os pesos utilizando os flags determinados
                weights = create_feathering_weight(chip_size, chip_size, overlap,
                                                   feather_top=feather_top,
                                                   feather_bottom=feather_bottom,
                                                   feather_left=feather_left,
                                                   feather_right=feather_right)

                patches.append((chip, x, y, valid_width, valid_height, weights, src.window_transform(win)))

    return patches, meta



def predict_on_patches(model, patches, mean, std):
    """Executa predições nas janelas mantendo os logits originais."""
    model.eval()
    transform = create_transforms(mean, std)
    predictions = []

    with torch.no_grad():
        for chip, x, y, w, h, weights, transform_geo in tqdm(patches, desc="Predicting", leave=False):
            input_tensor = torch.from_numpy(chip)
            input_tensor = transform(input_tensor)

            sample = {
                "pixels": input_tensor.unsqueeze(0),
                "time": torch.zeros(1, 4),
                "latlon": torch.zeros(1, 4),
            }
            sample = {k: v.to("cuda") if torch.is_tensor(v) else v for k, v in sample.items()}

            # Obtém logits brutos do modelo
            logits = model(sample)
            #logits = F.interpolate(logits, size=(h, w), mode="bilinear", align_corners=False)

            predictions.append((
                logits.squeeze(0).cpu().numpy(),  # (C, H, W)
                x, y, w, h,
                weights,
                transform_geo
            ))

    return predictions

def reconstruct_geotiff(predictions, meta, output_path, original_height, original_width, chip_size, return_probs=False, prob_class=None):
    """Reconstrói a imagem com blending suave, evitando artefatos nas bordas e removendo o padding."""
    height = meta['height']
    width = meta['width']
    num_classes = predictions[0][0].shape[0]

    # Inicializa acumuladores
    accumulated_logits = np.zeros((num_classes, height, width), dtype=np.float32)
    weight_sum = np.zeros((height, width), dtype=np.float32)

    # Processa cada predição
    for logits, x, y, w, h, weights, _ in predictions:
        # Normaliza pesos para evitar que bordas tenham menor influência
        weights = weights / np.max(weights)

        # Aplica pesos apenas na área válida
        valid_weights = np.zeros((chip_size, chip_size), dtype=np.float32)
        valid_weights[:h, :w] = weights[:h, :w]

        for c in range(num_classes):
            accumulated_logits[c, y:y+h, x:x+w] += logits[c, :h, :w] * valid_weights[:h, :w]
        weight_sum[y:y+h, x:x+w] += valid_weights[:h, :w]

    # Remove o padding da imagem reconstruída e do weight_sum
    final_logits = accumulated_logits[:, :original_height, :original_width]
    weight_sum = weight_sum[:original_height, :original_width]

    # Evita divisão por zero
    mask = weight_sum > 0
    final_logits[:, mask] /= weight_sum[mask]

    # Converte para probabilidades
    probs = torch.softmax(torch.from_numpy(final_logits), dim=0).numpy()

    # Prepara a saída final
    if return_probs:
        if prob_class is not None:
            if isinstance(prob_class, int):
                output = probs[prob_class:prob_class+1]  # Mantém a dimensão (1, H, W)
            else:
                output = probs[list(prob_class)]  # Seleciona múltiplas classes
        else:
            output = probs  # Todas as classes
        dtype = rasterio.float32
    else:
        output = np.argmax(probs, axis=0).astype(np.uint8)  # Máscara final
        dtype = rasterio.uint8

    # Atualiza metadados
    meta.update({
        'count': output.shape[0] if return_probs else 1,
        'dtype': dtype,
        'driver': 'GTiff',
        'compress': 'lzw',
        'height': original_height,
        'width': original_width
    })

    # Salva o resultado
    with rasterio.open(output_path, 'w', **meta) as dst:
        if return_probs:
            dst.write(output)  # Escreve todas as bandas de uma vez
        else:
            dst.write(output, 1)  # Escreve a máscara final




def run_geotiff_prediction(model, image_path, output_path, mean, std, chip_size=256, overlap=64, return_probs=False, prob_class=None):
    """Função principal de execução com parâmetros ajustados."""
    with rasterio.open(image_path) as src:
        original_height, original_width = src.height, src.width

    patches, meta = split_geotiff(image_path, chip_size, overlap)
    predictions = predict_on_patches(model, patches, mean, std)
    reconstruct_geotiff(predictions, meta, output_path, original_height, original_width, chip_size, return_probs, prob_class)

def batch_predict_on_directory(model, input_dir, output_dir, mean, std, chip_size=256, overlap=32, return_probs=False, prob_class=None):
    """
    Executa a predição em todos os arquivos GeoTIFF de um diretório.

    Args:
        model: Modelo de segmentação.
        input_dir: Diretório contendo os arquivos de entrada.
        output_dir: Diretório onde salvar os resultados.
        mean: Média para normalização.
        std: Desvio padrão para normalização.
        chip_size: Tamanho dos chips para segmentação.
        overlap: Sobreposição entre os chips.
        return_probs: Se True, exporta a probabilidade da classe 1.
    """
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    tiff_files = list(input_dir.glob("*.tif"))
    for image_path in tqdm(tiff_files, desc="Processando arquivos GeoTIFF"):
        output_path = output_dir / f"{image_path.stem}_pred.tif"
        run_geotiff_prediction(model, image_path, output_path, mean, std, chip_size, overlap, return_probs, prob_class)
```

### Funções para predição de problema binários

```python
import os
from pathlib import Path
import rasterio
from rasterio.windows import Window
import numpy as np
import torch
import torch.nn.functional as F
from tqdm import tqdm
from torchvision.transforms import v2

def get_model(chesapeake_checkpoint_path, clay_checkpoint_path, metadata_path):
    model = ChesapeakeSegmentor.load_from_checkpoint(
        checkpoint_path=chesapeake_checkpoint_path,
        metadata_path=metadata_path,
        ckpt_path=clay_checkpoint_path,
    )
    model.eval()
    return model

def create_transforms(mean, std):
    return v2.Compose([
        v2.Normalize(mean=mean, std=std),
    ])

def create_feathering_weight(height, width, overlap, feather_top=True, feather_bottom=True, feather_left=True, feather_right=True):
    """
    Cria uma matriz de pesos para blending suave. Se o feathering estiver
    desabilitado para um lado, os pesos nessa direção serão forçados a 1.
    """
    # Cria arrays para as coordenadas
    y = np.arange(height).reshape(-1, 1)
    x = np.arange(width).reshape(1, -1)

    # Se não for aplicar feathering, force a distância a ser igual a overlap (garantindo peso 1)
    top = y if feather_top else np.full((height, 1), overlap, dtype=np.int32)
    bottom = (height - y - 1) if feather_bottom else np.full((height, 1), overlap, dtype=np.int32)
    left = x if feather_left else np.full((1, width), overlap, dtype=np.int32)
    right = (width - x - 1) if feather_right else np.full((1, width), overlap, dtype=np.int32)

    # Calcula a distância mínima em relação a todas as bordas
    min_dist = np.minimum(np.minimum(top, bottom), np.minimum(left, right))

    # Converte essa distância em pesos lineares na região de overlap (0 a 1)
    weights = np.clip(min_dist / overlap, 0.0, 1.0).astype(np.float32)
    return weights

def compute_window_coords(image_dim, chip_size, stride):
    """Garante que a última janela esteja alinhada à borda."""
    coords = [] if image_dim > chip_size else [0]
    pos = 0
    while True:
        if pos + chip_size >= image_dim:
            coords.append(image_dim - chip_size)
            break
        else:
            coords.append(pos)
            pos += stride
    return coords

def split_geotiff(image_path, chip_size, overlap):
    """Divide a imagem em patches, garantindo alinhamento correto e ajustando os pesos nas bordas externas."""
    with rasterio.open(image_path) as src:
        width, height = src.width, src.height
        meta = src.meta.copy()
        patches = []
        stride = chip_size - overlap

        x_coords = compute_window_coords(width, chip_size, stride)
        y_coords = compute_window_coords(height, chip_size, stride)

        for y in y_coords:
            for x in x_coords:
                win = Window(x, y, chip_size, chip_size)
                chip = src.read(window=win, boundless=True, fill_value=0).astype(np.float32)

                valid_width = min(chip_size, width - x)
                valid_height = min(chip_size, height - y)

                # Define flags para feathering:
                # Se o patch estiver na borda externa, não aplica feathering naquele lado (força peso = 1).
                feather_top = False if y == 0 else True
                feather_left = False if x == 0 else True
                feather_bottom = False if (y + chip_size) >= height else True
                feather_right = False if (x + chip_size) >= width else True

                weights = create_feathering_weight(chip_size, chip_size, overlap,
                                                   feather_top=feather_top,
                                                   feather_bottom=feather_bottom,
                                                   feather_left=feather_left,
                                                   feather_right=feather_right)

                patches.append((chip, x, y, valid_width, valid_height, weights, src.window_transform(win)))

    return patches, meta

def predict_on_patches(model, patches, mean, std):
    """Executa predições e retorna os logits (sem aplicar sigmoid) para cada patch."""
    model.eval()
    transform = create_transforms(mean, std)
    predictions = []

    with torch.no_grad():
        for chip, x, y, w, h, weights, transform_geo in tqdm(patches, desc="Predicting", leave=None):
            input_tensor = torch.from_numpy(chip)
            input_tensor = transform(input_tensor)

            sample = {
                "pixels": input_tensor.unsqueeze(0),
                "time": torch.zeros(1, 4),
                "latlon": torch.zeros(1, 4),
            }
            sample = {k: v.to("cuda") if torch.is_tensor(v) else v for k, v in sample.items()}

            logits = model(sample)  # Saída: (1, 1, H, W)
            logits = logits.squeeze(0).squeeze(0).cpu().numpy()  # Resultado: (H, W)
            predictions.append((logits, x, y, w, h, weights, transform_geo))

    return predictions

def reconstruct_geotiff(predictions, meta, output_path, original_height, original_width, chip_size, return_probs=False):
    """
    Reconstrói a imagem final usando blending suave.
    Para problemas binários, a saída é processada com sigmoid.
    """
    height, width = meta['height'], meta['width']

    accumulated_logits = np.zeros((height, width), dtype=np.float32)
    weight_sum = np.zeros((height, width), dtype=np.float32)

    for logits, x, y, w, h, weights, _ in predictions:
        # Normaliza os pesos do patch para que o valor máximo seja 1
        weights = weights / np.max(weights)

        valid_weights = np.zeros((chip_size, chip_size), dtype=np.float32)
        valid_weights[:h, :w] = weights[:h, :w]

        accumulated_logits[y:y+h, x:x+w] += logits[:h, :w] * valid_weights[:h, :w]
        weight_sum[y:y+h, x:x+w] += valid_weights[:h, :w]

    final_logits = accumulated_logits[:original_height, :original_width]
    weight_sum = weight_sum[:original_height, :original_width]

    mask = weight_sum > 0
    final_logits[mask] /= weight_sum[mask]

    # Agora, aplicamos sigmoid somente na saída final (para obter probabilidades entre 0 e 1)
    probs = torch.sigmoid(torch.from_numpy(final_logits)).numpy()

    if return_probs:
        output = probs.astype(np.float32)  # Mantém as probabilidades
        dtype = rasterio.float32
    else:
        output = (probs > 0.5).astype(np.uint8)  # Binariza (0 ou 1)
        dtype = rasterio.uint8

    meta.update({
        'count': 1,
        'dtype': dtype,
        'driver': 'GTiff',
        'compress': 'lzw',
        'height': original_height,
        'width': original_width
    })

    with rasterio.open(output_path, 'w', **meta) as dst:
        dst.write(output, 1)

def run_geotiff_prediction(model, image_path, output_path, mean, std, chip_size=256, overlap=64, return_probs=False):
    """Executa a inferência em um único arquivo GeoTIFF."""
    with rasterio.open(image_path) as src:
        original_height, original_width = src.height, src.width

    patches, meta = split_geotiff(image_path, chip_size, overlap)
    predictions = predict_on_patches(model, patches, mean, std)
    reconstruct_geotiff(predictions, meta, output_path, original_height, original_width, chip_size, return_probs)

def batch_predict_on_directory(model, input_dir, output_dir, mean, std, chip_size=256, overlap=32, return_probs=False):
    """
    Executa a inferência em todos os arquivos GeoTIFF de um diretório.
    """
    input_dir = Path(input_dir)
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    tiff_files = list(input_dir.glob("*.tif"))
    for image_path in tqdm(tiff_files, desc="Processando arquivos GeoTIFF", leave=None):
        output_path = output_dir / f"{image_path.stem}_pred.tif"
        run_geotiff_prediction(model, image_path, output_path, mean, std, chip_size, overlap, return_probs)
```
